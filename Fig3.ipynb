# Fig 3 Histograms

This notebook generates all histogram plots included in figure 3.

The dataset used for constructing the plots in Fig 3 are stored in `data/ExperimentRepresentationEvaluation`. This dataset can be reproduced by running

```python -m main -e ExperimentSetRepresentationEvaluation```

from the project's root directory. Note that the plotting functions used in this notebook randomly decide which task of the given task set is used as a transfer task and which task is used for dermining the top 5\% scoring state abstractions.

````python
import warnings; warnings.filterwarnings('ignore')

from matplotlib import rc
rc('font', **{'family': 'serif', 'size': 9})
rc('text', usetex=True)

import rewardpredictive as rp
import matplotlib.pyplot as plt

rp.set_seeds(12345)
experiment_set = rp.ExperimentSetRepresentationEvaluation.load()
````

### Column World Task Set

````python
exp_column_world = experiment_set.get_experiment_by_task_sequence_name(rp.SmallTaskSequenceName.COLUMN_WORLD)
rp.plot_and_save_histogram_column_world('figure', exp_column_world)
````

### Random MDP Task Set

````python
exp_rand_mdp = experiment_set.get_experiment_by_task_sequence_name(rp.SmallTaskSequenceName.RAND_MDP)
rp.plot_and_save_histogram_rand_mdp('figure', exp_rand_mdp)
````

### Grid World Task Set

````python
exp_grid_world = experiment_set.get_experiment_by_task_sequence_name(rp.SmallTaskSequenceName.GOAL_WALL_GRID)
rp.plot_and_save_histogram_goal_and_wall_world('figure', exp_grid_world)
````
